```python
import sys
import os
sys.path.append(os.path.realpath('..'))
from apriori_implementation import apriori
import pandas
import numpy as np
import time
```

```python
file_path = os.path.realpath('..') + '/resources/t25i10d10_transformed.txt'
```

```python
file_path
```

```python
raw = pandas.read_csv(file_path, header=None)
```

```python
raw
```

```python
def transform(row):
    return row.strip().replace(' ', ',')

transactions = raw[0].apply(transform)
```

```python
transactions
```

```python
dataset_size = transactions.shape[0]
```

```python
transactions_list = []
for i in range(dataset_size):
    transactions_list.append(transactions[i])
```

```python
transactions_list
```

```python
data = {'col': transactions_list}
df = pandas.DataFrame.from_dict(data)
```

```python
df
```

```python
start = time.time()

apriori(df, 0.8, 0.95)

end = time.time()
print('\nExecuÃ§Ã£o durou', end - start, 'segundos')
```
